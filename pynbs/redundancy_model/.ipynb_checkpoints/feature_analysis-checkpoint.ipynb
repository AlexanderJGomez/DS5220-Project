```python
import sqlite3
import pandas as pd
import itertools
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr, spearmanr
```

```python
conn = sqlite3.connect("../../data/Stanford_Data/cdm_postedits.db")
df = pd.read_sql("SELECT * FROM cdm_postedits", conn)
conn.close()
```

## Analysis of ELMo/Universal Sentence Encoder Redundancy Scores and Human Redundancy Scores

```python
avg_redundancy_scores = []
_unique_vals, indices = np.unique([row.system + str(row.id) for _, row in df.iterrows()], return_index=True)
avg_redundancy_df = df.iloc[indices].sort_index()
for _index, row in avg_redundancy_df.iterrows():
    all_scores = df[(row.id == df.id) & (row.system == df.system)].redundancy
    avg_redundancy_scores.append(np.mean(all_scores))
```

#### Universal Sentence Encoder

For Mean

```python
pearsonr(avg_redundancy_scores, avg_redundancy_df.mean_use_redundancy)
```

```python
spearmanr(avg_redundancy_scores, avg_redundancy_df.mean_use_redundancy)
```

```python
plt.ylabel("Univ. Sentence Encoder Score")
plt.xlabel("Avg Human Redundancy Score")
plt.scatter(list(avg_redundancy_scores), avg_redundancy_df.mean_use_redundancy, s=10, alpha=0.05)
plt.show()
```

```python
use_redundany_score_pairs = list(zip(avg_redundancy_scores, avg_redundancy_df.mean_use_redundancy))
unique_scores = np.unique(avg_redundancy_scores)
for score in unique_scores:
    use_scores = [x[1] for x in use_redundany_score_pairs if x[0] == score]
    if len(use_scores) > 3:
        print("For score " + str(score) + ": mean is %0.3f and SD is %0.3f" %
         (np.mean(use_scores), np.std(use_scores)))
```

## Pearson Correlations per Annotator
This sections contains correlations between ELMo and Universal Sentence Encoder Redundancy scores and human redundancy scores for each annotator.  If there is an annotator for which the correlation is zero it is because they annotated less than 3 system summaries and we assign a zero correlation by default because we cannot gain any information with less than 3 data points.

```python
annotator_ids = np.unique(df.annotator_id)
annotator_use_correlations = []
annotator_elmo_correlations = []
for annotator_id in annotator_ids:
    annotator_rows = df[df.annotator_id == annotator_id]
    if annotator_rows.shape[0] < 3:
        annotator_use_correlations.append((0, 1))
        annotator_elmo_correlations.append((0, 1))
        continue
    use_pearson = pearsonr(annotator_rows.max_use_redundancy, annotator_rows.redundancy)
    elmo_pearson = pearsonr(annotator_rows.max_elmo_redundancy, annotator_rows.redundancy)
    annotator_use_correlations.append(use_pearson)
    annotator_elmo_correlations.append(elmo_pearson)
    
```

```python
zipped_id_correlation_use = list(zip(annotator_ids, annotator_use_correlations))
zipped_id_correlation_elmo = list(zip(annotator_ids, annotator_elmo_correlations))
significant_correlations_use = [x for x in zipped_id_correlation_use if x[1][1] < 0.05]
significant_correlations_elmo = [x for x in zipped_id_correlation_elmo if x[1][1] < 0.05]
```

### Histogram for significant annotator correlations (p < 0.05)

This shows a histogram of the Pearson correlation scores that occurred between each metric and an annotator's scores.  It only includes scores if the p value was < 0.05

```python
plt.hist(x=[x[1][0] for x in significant_correlations_use], edgecolor='black')
plt.xlabel("Pearson r")
plt.ylabel("Number of Annotators")
plt.title("For Universal Sentence Encoder")
plt.show()
```

```python
plt.hist(x=[x[1][0] for x in significant_correlations_elmo], edgecolor='black')
plt.xlabel("Pearson r")
plt.ylabel("Number of Annotators")
plt.title("For ELMo")
plt.show()
```
