```python
import sqlite3
import pandas as pd
import itertools
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr, spearmanr
```

```python
conn = sqlite3.connect("../data/cdm_postedits.db")
df = pd.read_sql("SELECT * FROM cdm_postedits", conn)
conn.close()
```

## Analysis of ELMo/Universal Sentence Encoder Redundancy Scores and Human Redundancy Scores

#### Universal Sentence Encoder

```python
plt.ylabel("Univ. Sentence Encoder Score")
plt.xlabel("Human Redundancy Score")
plt.scatter(list(df.redundancy), df.use_redundancy, s=10, alpha=0.01)
plt.show()
```

```python
print("Universal Sentence Encoder Pearson Correlation score: " + str(pearsonr(list(df.redundancy), df.use_redundancy)))
```

```python
sup = list(zip(df.redundancy, df.use_redundancy))
bad_avg = [x[1] for x in sup if x[0] == -1]
normal_avg = [x[1] for x in sup if x[0] == 0]
good_avg = [x[1] for x in sup if x[0] == 1]
print("redundancy score -1:  mean is %0.3f, SD is %0.3f" % ((sum(bad_avg) / len(bad_avg)), np.std(bad_avg)))
print("redundancy score 0: mean is %0.3f, SD is %0.3f" % ((sum(normal_avg) / len(normal_avg)), np.std(normal_avg)))
print("redundancy score 1: mean is %0.3f, SD is %0.3f" % ((sum(good_avg) / len(good_avg)), np.std(good_avg)))
```

#### ELMo

```python
plt.ylabel("ELMo Score")
plt.xlabel("Human Redundancy Score")
plt.scatter(list(df.redundancy), df.elmo_redundancy, s=10, alpha=0.01)
plt.show()
```

```python
print("ELMo Embedding Pearson Correlation score: " + str(pearsonr(list(df.redundancy), df.elmo_redundancy)))
```

```python
elmo_averages = list(zip(df.redundancy, df.elmo_redundancy))
elmo_bad = [x[1] for x in elmo_averages if x[0] == -1]
elmo_average = [x[1] for x in elmo_averages if x[0] == 0]
elmo_good = [x[1] for x in elmo_averages if x[0] == 1]
print("for redundancy score -1: mean is %0.3f, SD is %0.3f" % ((sum(elmo_bad) / len(elmo_bad)), np.std(elmo_bad)))
print("for redundancy score 0: mean is %0.3f, SD is %0.3f" % ((sum(elmo_average) / len(elmo_average)), np.std(elmo_average)))
print("for redundancy score 1: mean is %0.3f, SD is %0.3f" % ((sum(elmo_good) / len(elmo_good)), np.std(elmo_good)))
```

## Analysis for Average Redundancy Score per System Summary

This is if we average the human evaluation score per system summary, rather than treating them separately as we did above

```python
avg_redundancy_scores = []
_unique_vals, indices = np.unique([row.system + str(row.id) for _, row in df.iterrows()], return_index=True)
avg_redundancy_df = df.iloc[indices].sort_index()
for _index, row in avg_redundancy_df.iterrows():
    all_scores = df[(row.id == df.id) & (row.system == df.system)].redundancy
    avg_redundancy_scores.append(np.mean(all_scores))
```

#### Universal Sentence Encoder

```python
pearsonr(avg_redundancy_scores, avg_redundancy_df.use_redundancy)
```

```python
plt.ylabel("Univ. Sentence Encoder Score")
plt.xlabel("Avg Human Redundancy Score")
plt.scatter(list(avg_redundancy_scores), avg_redundancy_df.use_redundancy, s=10, alpha=0.05)
plt.show()
```

```python
use_redundany_score_pairs = list(zip(avg_redundancy_scores, avg_redundancy_df.use_redundancy))
unique_scores = np.unique(avg_redundancy_scores)
for score in unique_scores:
    use_scores = [x[1] for x in use_redundany_score_pairs if x[0] == score]
    if len(use_scores) > 3:
        print("For score " + str(score) + ": mean is %0.3f and SD is %0.3f" %
         (np.mean(use_scores), np.std(use_scores)))
```

#### ELMo

```python
pearsonr(avg_redundancy_scores, avg_redundancy_df.elmo_redundancy)
```

```python
plt.ylabel("ELMo Score")
plt.xlabel("Human Redundancy Score")
plt.scatter(list(avg_redundancy_scores), avg_redundancy_df.elmo_redundancy, s=10, alpha=0.05)
plt.show()
```

```python
elmo_redundany_score_pairs = list(zip(avg_redundancy_scores, avg_redundancy_df.elmo_redundancy))
unique_scores = np.unique(avg_redundancy_scores)
for score in unique_scores:
    elmo_scores = [x[1] for x in elmo_redundany_score_pairs if x[0] == score]
    if len(elmo_scores) > 3:
        print("For score " + str(score) + ": mean is %0.3f and SD is %0.3f" %
         (np.mean(elmo_scores), np.std(elmo_scores)))
```

## Pearson Correlations per Annotator
This sections contains correlations between ELMo and Universal Sentence Encoder Redundancy scores and human redundancy scores for each annotator.  If there is an annotator for which the correlation is zero it is because they annotated less than 3 system summaries and we assign a zero correlation by default because we cannot gain any information with less than 3 data points.

```python
annotator_ids = np.unique(df.annotator_id)
annotator_use_correlations = []
annotator_elmo_correlations = []
for annotator_id in annotator_ids:
    annotator_rows = df[df.annotator_id == annotator_id]
    if annotator_rows.shape[0] < 3:
        annotator_use_correlations.append((0, 1))
        annotator_elmo_correlations.append((0, 1))
        continue
    use_pearson = pearsonr(annotator_rows.use_redundancy, annotator_rows.redundancy)
    elmo_pearson = pearsonr(annotator_rows.elmo_redundancy, annotator_rows.redundancy)
    annotator_use_correlations.append(use_pearson)
    annotator_elmo_correlations.append(elmo_pearson)
    
```

#### Universal Sentence Encoder Correlations by Annotator ID

```python
zipped_id_correlation_use = list(zip(annotator_ids, annotator_use_correlations))
zipped_id_correlation_elmo = list(zip(annotator_ids, annotator_elmo_correlations))
significant_correlations_use = [x for x in zipped_id_correlation_use if x[1][1] < 0.05]
significant_correlations_elmo = [x for x in zipped_id_correlation_elmo if x[1][1] < 0.05]
for x in zipped_id_correlation_use:
    print("id: %s  correlation: %0.3f (p<%s)" % (x[0], x[1][0], x[1][1]))
```

#### ELMo Correlations by Annotator ID

```python
for x in zipped_id_correlation_elmo:
    print("id: %s  correlation: %0.3f (p<%s)" % (x[0], x[1][0], x[1][1]))
```

### Histogram for significant annotator correlations (p < 0.05)

This shows a histogram of the Pearson correlation scores that occurred between each metric and an annotator's scores.  It only includes scores if the p value was < 0.05

```python
plt.hist(x=[x[1][0] for x in significant_correlations_use], edgecolor='black')
plt.xlabel("Pearson r")
plt.ylabel("Number of Annotators")
plt.title("For Universal Sentence Encoder")
plt.show()
```

```python
plt.hist(x=[x[1][0] for x in significant_correlations_elmo], edgecolor='black')
plt.xlabel("Pearson r")
plt.ylabel("Number of Annotators")
plt.title("For ELMo")
plt.show()
```
