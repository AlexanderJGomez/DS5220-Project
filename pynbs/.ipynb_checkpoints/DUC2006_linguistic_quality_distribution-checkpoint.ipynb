```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py

df = pd.read_table("../../data/NISTeval/linguistic_quality/linguistic_quality.table", skiprows=7,
              names=["topic","selector", "assessor", "system", "qqn", "score"])
baseline_df = df[df.system == "1"]
humans_df = df[[x.isalpha() for x in df.system]]
systems_df = df[[x.isdigit() and x != "1" for x in df.system]]
```

Grammaticality Distribution

```python
width = 1/1.5
vals, counts = np.unique(humans_df[humans_df.qqn == 1].score, return_counts=True)
plt.figure(1)
plt.bar(vals, counts, width, color="blue")
plt.xlabel("Human")
plt.ylabel("Count")
plt.figure(2)
vals, counts = np.unique(baseline_df[baseline_df.qqn == 1].score, return_counts=True)
plt.bar(vals, counts, width, color="blue")
plt.xlabel("Baseline")
plt.ylabel("Count")
plt.figure(3)
vals, counts = np.unique(systems_df[systems_df.qqn == 1].score, return_counts=True)
plt.bar(vals, counts, width, color="blue")
plt.xlabel("System")
plt.ylabel("Count")
plt.show()
```
